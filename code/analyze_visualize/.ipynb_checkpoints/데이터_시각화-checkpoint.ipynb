# Data Pre-Processing
    주가 데이터에 대해 날짜 인덱스로 변환
    주가 데이터와 재무제표 데이터간 시제 통일 (년도 단위)
        SM => 2014 ~ 2022
        JYP => 2014 ~ 2022
        YG => 2014 ~ 2022

```python
sm_stock['날짜']=pd.to_datetime(sm_stock['날짜'])
sm_stock.sort_values(by='날짜',inplace=True)
sm_stock.reset_index(drop=True,inplace=True)
sm_stock.columns=['', '종가', '오픈', '고가', '저가', '거래량','변동 %']
sm_stock.set_index('',inplace=True)

jyp_stock['날짜']=pd.to_datetime(jyp_stock['날짜'])
jyp_stock.sort_values(by='날짜',inplace=True)
jyp_stock.reset_index(drop=True,inplace=True)
jyp_stock.columns=['', '종가', '오픈', '고가', '저가', '거래량','변동 %']
jyp_stock.set_index('',inplace=True)

yg_stock['날짜']=pd.to_datetime(yg_stock['날짜'])
yg_stock.sort_values(by='날짜',inplace=True)
yg_stock.reset_index(drop=True,inplace=True)
yg_stock.columns=['', '종가', '오픈', '고가', '저가', '거래량','변동 %']
yg_stock.set_index('',inplace=True)
```

```python
# SM 년도별 평균 주가
sm_stock_year_mean=sm_stock.resample('Y').mean()
sm_stock_year_mean.reset_index(inplace=True)
sm_stock_year_mean['']=pd.to_datetime(sm_stock_year_mean['']).dt.year
sm_stock_year_mean.set_index('',inplace=True)
sm_stock_year_mean.drop([2023],inplace=True)
sm_stock_year_mean
```

```python
sm_chart.drop(2013,inplace=True)
sm_chart
```

```python
# JYP 년도별 평균 주가
jyp_stock_year_mean=jyp_stock.resample('Y').mean()
jyp_stock_year_mean.reset_index(inplace=True)
jyp_stock_year_mean['']=pd.to_datetime(jyp_stock_year_mean['']).dt.year
jyp_stock_year_mean.set_index('',inplace=True)
jyp_stock_year_mean.drop([2023],inplace=True)
jyp_stock_year_mean
```

```python
jyp_chart.drop(2013,inplace=True)
```

```python
# YG 년도별 평균 주가
yg_stock_year_mean=yg_stock.resample('Y').mean()
yg_stock_year_mean.reset_index(inplace=True)
yg_stock_year_mean['']=pd.to_datetime(yg_stock_year_mean['']).dt.year
yg_stock_year_mean.set_index('',inplace=True)
yg_stock_year_mean.drop([2023],inplace=True)
yg_stock_year_mean
```

```python
yg_chart.drop(2013,inplace=True)
```

```python
plt.figure(figsize=(16,9))
for i in range(8):
    if (i+1)%2 ==1:
        plt.subplot(4,2,i+1)
        sns.lineplot(data=sm_stock_year_mean,x=sm_stock_year_mean.index,y='종가')
        plt.title('SM 년 평균 주가')
    else:
        plt.subplot(4,2,i+1)
        sns.lineplot(data=sm_chart,x=sm_chart.index,y=sm_chart.columns[i//2])
        plt.title(f'SM {sm_chart.columns[i//2]}')
plt.tight_layout()
plt.show()
```
