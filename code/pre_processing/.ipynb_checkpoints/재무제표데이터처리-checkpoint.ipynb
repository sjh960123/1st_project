# 재무제표 데이터 처리

```python
import pandas as pd
import numpy as np
import os
```

## 재무제표 이름 재설정
    ex) [에스엠]사업보고서_재무제표(2015.03.31)_ko.xls => [에스엠]2015.xls

```python
# 각 기업별 재무제표 데이터 경로 추출
data_dir_path='D:/이직준비공부/AI/project_1/data/재무제표'

for file in os.listdir(data_dir_path):
    year=file.split('.')[0].split('(')[-1]
    if file.startswith('[에스엠]'):
        os.rename(os.path.join(data_dir_path,file),os.path.join(data_dir_path,'[에스엠]'+year+'.xls'))
        
    elif file.startswith('[JYP Ent]'):
        os.rename(os.path.join(data_dir_path,file),os.path.join(data_dir_path,'[JYP Ent]'+year+'.xls'))
        
    else:
        os.rename(os.path.join(data_dir_path,file),os.path.join(data_dir_path,'[와이지엔터테인먼트]'+year+'.xls'))
```

## 재무제표 데이터 통합
    각 기업의 2013년도 부터 공시된 최근 재무제표 데이터 통합

```python
# 각 기업별 재무제표 데이터 경로 추출
data_dir_path='D:/이직준비공부/AI/project_1/data/재무제표'
sm_list=[]
jyp_list=[]
yg_list=[]

for file in os.listdir(data_dir_path):

    if file.startswith('[에스엠]'):
        sm_list.append(os.path.join(data_dir_path,file))
        
    elif file.startswith('[JYP Ent]'):
        jyp_list.append(os.path.join(data_dir_path,file))
        
    else:
        yg_list.append(os.path.join(data_dir_path,file))
```

### YG Ent 재무제표 통합

### SM Ent 재무제표

```python
for i,path in enumerate(sm_list):
    # 각 엑셀 별 데이터프레임 생성
    globals()["sm_df_{}".format(i)]=pd.read_excel(path,sheet_name='재무상태표',header=6)
    
    # 항목에 대해 공백제거 처리
    globals()["sm_df_{}".format(i)][' ']=globals()["sm_df_{}".format(i)][' '].str.strip()
    globals()["sm_df_{}".format(i)][' ']=globals()["sm_df_{}".format(i)][' '].apply(lambda x: x.split('(')[0])
    globals()["sm_df_{}".format(i)][' ']=globals()["sm_df_{}".format(i)][' '].apply(lambda x: x.split(' ')[0])
    # 필요 항목 추출
    drop_list=[i for i in globals()["sm_df_{}".format(i)][' '].values if i not in ['매출채권','재고자산','투자부동산','무형자산','자본금','이익잉여금']]
    
    globals()["sm_df_{}".format(i)].set_index(' ',inplace=True)
    
    globals()["sm_df_{}".format(i)].drop(drop_list,axis=0,inplace=True)
```

```python
# 데이터프레임 통합
sm_df=pd.concat([sm_df_0,sm_df_1,sm_df_2,sm_df_3,sm_df_4,sm_df_5,sm_df_6,sm_df_7],axis=1).T
sm_df
```

```python
sm_df.reset_index(inplace=True)
sm_df
```

```python
# 의미없는 기수 제거
sm_df.drop([3,4,5],axis=0,inplace=True)
sm_df
```

```python
# 중복되는 기수 제거
sm_df.drop_duplicates('index',inplace=True)
sm_df
```

```python
# 테이터 프레임 
sm_df.columns=['','매출채권','재고자산','무형자산','자본금','이익잉여금','투자부동산']
sm_df.set_index('',inplace=True)
sm_df=sm_df.T
sm_df
```

```python
sm_df.to_csv('D:/이직준비공부/AI/project_1/data/재무제표_통합/SM_재무제표.csv')
```
