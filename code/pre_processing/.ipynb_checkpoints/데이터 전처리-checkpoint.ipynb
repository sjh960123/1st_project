```python
import pandas as pd
import numpy as np
```

## Data Load

```python
GDP_df=pd.read_csv('D:/이직준비공부/AI/project_1/data/경제지표_데이터/GDP_2013_2022.csv',encoding='ms949')
interset_rate_df=pd.read_csv('D:/이직준비공부/AI/project_1/data/경제지표_데이터/한국은행 기준금리 및 여수신금리_17192537.csv')
exchage_rate_df=pd.read_csv('D:/이직준비공부/AI/project_1/data/경제지표_데이터/환율.csv')

sm_stock=pd.read_csv('D:/이직준비공부/AI/project_1/data/주가_데이터/SM_Ent.csv')
jyp_stock=pd.read_csv('D:/이직준비공부/AI/project_1/data/주가_데이터/JYP_Ent.csv')
yg_stock=pd.read_csv('D:/이직준비공부/AI/project_1/data/주가_데이터/YG_Ent.csv')

kosdex=pd.read_csv('D:/이직준비공부/AI/project_1/data/주가_데이터/KOSDEX.csv')

sm_chart=pd.read_csv('D:/이직준비공부/AI/project_1/data/재무제표_통합/SM_재무제표.csv')
jyp_chart=pd.read_csv('D:/이직준비공부/AI/project_1/data/재무제표_통합/JYP_재무제표.csv')
yg_chart=pd.read_csv('D:/이직준비공부/AI/project_1/data/재무제표_통합/YG_재무제표.csv')
```

## GDP 데이터 처리

```python
# 특정 행 추출 (1인 GDP)
GDP_df=GDP_df[GDP_df['계정항목별']=='1인당 국내총생산(명목, 원화표시) (만원)']
GDP_df.reset_index(drop=True,inplace=True)
GDP_df
```

```python
# 데이터프레임 재설정
GDP_df.loc[0,'계정항목별']='GDP'
GDP_df.columns=['', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020',
       '2021', '2022']
GDP_df.set_index('',inplace=True)
GDP_df
```

```python
GDP_df.info()
```

```python
# 타입 변경
GDP_df=GDP_df.T
GDP_df['GDP']=GDP_df['GDP'].astype(float)
```

```python
GDP_df.info()
```

```python
# 단위 변환(만원 => 원)
GDP_df=GDP_df*10000
GDP_df
```

```python
# 년도 인덱스 처리
GDP_df.reset_index(inplace=True)
GDP_df.columns=['year', 'GDP']
GDP_df['year']=pd.to_datetime(GDP_df['year']).dt.year
GDP_df.columns=['', 'GDP']
GDP_df.set_index('',inplace=True)
GDP_df
```

```python
GDP_df.to_csv('D:/이직준비공부/AI/project_1/data/전처리_완료_데이터/경제지표/GDP.csv')
```

## 금리 데이터 처리

```python
# 컬럼 명 변경
interset_rate_df.columns=['date', 'Rate(%)']
interset_rate_df
```

```python
# 데이터 타입 확인
interset_rate_df.info()
```

```python
# 날짜 데이터 타입 변경
interset_rate_df['date']=pd.to_datetime(interset_rate_df['date'])
interset_rate_df
```

```python
# 날짜 데이터 인덱스 지정
interset_rate_df.columns=['', 'Rate(%)']
interset_rate_df.set_index('',inplace=True)
interset_rate_df
```

```python
interset_rate_df.to_csv('D:/이직준비공부/AI/project_1/data/전처리_완료_데이터/경제지표/금리.csv')
```

## 환율 데이터

```python
exchage_rate_df
```

```python
# 데이터 타입 확인
exchage_rate_df.info()
```

```python
# 날짜 데이터 타입 변경
exchage_rate_df['날짜']=pd.to_datetime(exchage_rate_df['날짜'])
```

```python
exchage_rate_df.info()
```

```python
# 날짜 데이터 오름차순 정렬
exchage_rate_df.sort_values(by='날짜',inplace=True)
exchage_rate_df.reset_index(drop=True,inplace=True)
exchage_rate_df
```

```python
# 각 데이터에서 의미없는 특수 기호 제거
exchage_rate_df['종가']=exchage_rate_df['종가'].apply(lambda x: x.replace(',',''))
exchage_rate_df['오픈']=exchage_rate_df['오픈'].apply(lambda x: x.replace(',',''))
exchage_rate_df['고가']=exchage_rate_df['고가'].apply(lambda x: x.replace(',',''))
exchage_rate_df['저가']=exchage_rate_df['저가'].apply(lambda x: x.replace(',',''))
exchage_rate_df['변동 %']=exchage_rate_df['변동 %'].apply(lambda x: x.replace('%',''))
exchage_rate_df
```

```python
# 결측치가 많은 거래량 데이터 제거
exchage_rate_df.drop('거래량',axis=1,inplace=True)
exchage_rate_df
```

```python
# 각 데이터 타입 변경
exchage_rate_df['종가']=exchage_rate_df['종가'].astype(float)
exchage_rate_df['오픈']=exchage_rate_df['오픈'].astype(float)
exchage_rate_df['고가']=exchage_rate_df['고가'].astype(float)
exchage_rate_df['저가']=exchage_rate_df['저가'].astype(float)
exchage_rate_df['변동 %']=exchage_rate_df['변동 %'].astype(float)
exchage_rate_df
```

```python
exchage_rate_df.info()
```

```python
# 인덱스 지정
exchage_rate_df.columns=['', '종가', '오픈', '고가', '저가', '변동 %']
exchage_rate_df.set_index('',inplace=True)
exchage_rate_df
```

```python
exchage_rate_df.to_csv('D:/이직준비공부/AI/project_1/data/전처리_완료_데이터/경제지표/환율.csv')
```

## 코스닥 데이터 

```python
kosdex
```

```python
# 데이터 타입 확인
kosdex.info()
```

```python
# 날짜 데이터 타입 변경
kosdex['날짜']=pd.to_datetime(kosdex['날짜'])
```

```python
# 특수문자 제거
kosdex['종가']=kosdex['종가'].apply(lambda x: x.replace(',',''))
kosdex['오픈']=kosdex['오픈'].apply(lambda x: x.replace(',',''))
kosdex['고가']=kosdex['고가'].apply(lambda x: x.replace(',',''))
kosdex['저가']=kosdex['저가'].apply(lambda x: x.replace(',',''))
kosdex['변동 %']=kosdex['변동 %'].apply(lambda x: x.replace('%',''))
kosdex
```

```python
# 각 데이터 타입 변경
kosdex['종가']=kosdex['종가'].astype(float)
kosdex['오픈']=kosdex['오픈'].astype(float)
kosdex['고가']=kosdex['고가'].astype(float)
kosdex['저가']=kosdex['저가'].astype(float)
kosdex['변동 %']=kosdex['변동 %'].astype(float)
kosdex
```

```python
kosdex.info()
```

```python
# 거래량 데이터 처리
kosdex['거래량']=kosdex['거래량'].fillna('0')
kosdex['거래량']=kosdex['거래량'].apply(lambda x: float(x.split('M')[0])*1000000 if x.endswith('M') else float(x.split('K')[0])*1000)
kosdex
```

```python
kosdex.info()
```

```python
# 인덱스 지정
kosdex.columns=['', '종가', '오픈', '고가', '저가', '거래량','변동 %']
kosdex.set_index('',inplace=True)
kosdex
```

```python
kosdex.to_csv('D:/이직준비공부/AI/project_1/data/전처리_완료_데이터/주가/코스닥.csv')
```
