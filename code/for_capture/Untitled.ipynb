```python
# 각 기업별 재무제표 데이터 경로 추출
data_dir_path='D:/이직준비공부/AI/project_1/data/재무제표'
sm_list=[]
jyp_list=[]
yg_list=[]

for file in os.listdir(data_dir_path):

    if file.startswith('[에스엠]'):
        sm_list.append(os.path.join(data_dir_path,file))
        
    elif file.startswith('[JYP Ent]'):
        jyp_list.append(os.path.join(data_dir_path,file))
        
    else:
        yg_list.append(os.path.join(data_dir_path,file))
```

```python
for i,path in enumerate(yg_list):
    # 각 엑셀 별 데이터프레임 생성
    globals()["yg_df_{}".format(i)]=pd.read_excel(path,sheet_name='재무상태표',header=6)
    
    # 항목에 대해 공백제거 처리
    globals()["yg_df_{}".format(i)][' ']=globals()["yg_df_{}".format(i)][' '].str.strip()
    
    # 필요 항목 추출
    drop_list=[i for i in globals()["yg_df_{}".format(i)][' '].values if i not in ['매출채권','재고자산','투자부동산','무형자산','자본금','이익잉여금']]
    
    globals()["yg_df_{}".format(i)].set_index(' ',inplace=True)
    
    globals()["yg_df_{}".format(i)].drop(drop_list,axis=0,inplace=True)
```
